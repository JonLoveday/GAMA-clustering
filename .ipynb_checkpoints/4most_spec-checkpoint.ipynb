## Coadd joined spectra of same template type

```python
%cd '/research/astro/gama/loveday/Data_local/4most/OpR2'
cat = Table.read('/research/astro/gama/loveday/Data_sync/4MOST/OpR2exGalSample_170820_orig.csv')
# imin = np.argmin(cat['REDSHIFT'])
# print(cat[imin])
spectab = Table.read('spectab.fits')
fluxcon = FluxConservingResampler()
wmin, wmax = 3700, 9500  # Observed wavelength range
rng = np.random.default_rng()
nsum = 10  # Number of spectra to sum
for survey, tmplist in zip(('s7', 's8'), (('Pass', 'SF'), ('ELG', 'LRG', 'qso_BL'))):
    for tmp in tmplist:
        sel = (cat['SURVEY'] == survey) * (np.char.find(cat['TEMPLATE'], tmp) > -1)
        nobj = len(cat[sel])
        zmin = np.min(cat['REDSHIFT'])
        zmax = np.max(cat['REDSHIFT'])
        lmin, lmax = wmin/(1 + zmax), wmax/(1 + zmin)  # Restframe wavelength range
        disp_grid = np.arange(lmin, lmax, 3) * u.AA
        print(len(disp_grid), nsum)
        spec_list = []
#         specall = Spectrum1D(spectral_axis=disp_grid, 
#                              flux=np.zeros((nsum, len(disp_grid))) * u.erg / (u.Angstrom * u.cm**2 * u.s),
#                              uncertainty=StdDevUncertainty(np.zeros((nsum, len(disp_grid)))))
        flux = np.zeros((nsum, len(disp_grid)))
        ferr = np.zeros((nsum, len(disp_grid)))
        match = join(cat[sel], spectab, keys='CNAME', join_type='left')
        print(survey, tmp, len(cat[sel]), len(match), zmin, zmax, lmin, lmax)
        speclist = rng.choice(len(match), nsum)
        for i in range(nsum):
            ispec = speclist[i]
            specname = match['SPECNAME'][ispec]
            spect = Table.read(specname)
            flux_unit = spect['FLUX'].unit
#             print(spect['FLUX'].data[0], flux_unit)
            spec = Spectrum1D(spectral_axis=spect['WAVE'].data[0] * spect['WAVE'].unit / (1 + match['REDSHIFT'][ispec]), 
                              flux=spect['FLUX'].data[0] * flux_unit,
                              uncertainty=StdDevUncertainty(spect['ERR_FLUX'].data[0]))
#             spec = spectrum.Spectrum(disp=spect['WAVE'].data[0] * spect['WAVE'].unit / (1 + match['REDSHIFT'][ispec]), 
#                               flux=spect['FLUX'].data[0] * flux_unit,
#                               error=spect['ERR_FLUX'].data[0])
#             spec_list.append(spec)

            specr = fluxcon(spec, disp_grid)
            spec_list.append(specr)
#             specall.flux[i, :] = specr.flux
#            help(specr.uncertainty)
#             specall.uncertainty[i, :] = specr.uncertainty
            flux[i, :] = specr.flux
            ferr[i, :] = specr.uncertainty.array
#         spec_coadd = spectrum.coadd_errorweighted(spec_list, dispersion=disp_grid, bounds_error=False)
        plt.clf()
        indices = ~np.isnan(flux)
        print(ferr[indices])
#         plt.plot(spec_coadd.dispersion, spec_coadd.flux)
        plt.plot(disp_grid, np.average(flux, weights=ferr**-2, axis=0))

        plt.xlabel('Wavelength [A]')
        plt.ylabel(f'Flux [{flux_unit}]')
        plt.title(survey + tmp)
        plt.show()


```
